```python
import rclpy
from rclpy.node import Node
from geometry_msgs.msg import TwistStamped
from geometry_msgs.msg import PoseWithCovarianceStamped
from minibot_interfaces.msg import RobotState
from datetime import datetime
# import mysql.connector

class SubscriberNode(Node):

    def __init__(self):
        super().__init__('multi_topic_subscriber')

        self.create_subscription(
            TwistStamped,
            '/base_controller/cmd_vel_out',
            self.cmd_vel_out_callback,
            10)


        self.create_subscription(
            RobotState,
            '/minibot_io_controller/robot_state',
            self.robot_state_callback,
            10)

        self.subscription = self.create_subscription(
            PoseWithCovarianceStamped,
            '/amcl_pose',
            self.amcl_pose_callback,
            10)

    def cmd_vel_out_callback(self, msg):    
        timestamp_sec = msg.header.stamp.sec
        timestamp_nsec = msg.header.stamp.nanosec
        readable_timestamp = datetime.fromtimestamp(timestamp_sec + timestamp_nsec / 1e9).strftime('%Y-%m-%d %H:%M:%S.%f')
        print(readable_timestamp)
        cmd_linear = msg.twist.linear.x
        cmd_angular = msg.twist.angular.z

    def robot_state_callback(self, msg): 
        enable_motor = msg.enable_motor
        left_lamp = msg.left_lamp
        right_lamp = msg.right_lamp
   
    def amcl_pose_callback(self, msg):
        pose_x = msg.pose.pose.position.x
        pose_y = msg.pose.pose.position.y
        orient_z = msg.pose.pose.orientation.z
        orient_w = msg.pose.pose.orientation.w
       

def main(args=None):
    rclpy.init(args=args)
    node = SubscriberNode()

    try:
        rclpy.spin(node)
    except KeyboardInterrupt:
        pass

    node.destroy_node()
    rclpy.shutdown()

if __name__ == '__main__':
    main()
```
